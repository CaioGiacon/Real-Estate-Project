**DEVELOPMENT TEAM DEMANDS**

O time de desenvolvedores do site da empresa solicitou a criação de duas novas colunas numéricas na base de dados. No site dos imóveis, eles
gostariam de apresentar informações mais sumarizadas em relação a valores. Para isso, essas colunas serão importantes:

*   valor_por_mes: essa coluna deve conter os gastos mensais de cada imóvel, incluindo aluguel e condomínio;
*   valor_por_ano: essa coluna deve conter os gastos anuais por imóvel, ou seja, IPTU mais 12 meses de aluguel e condomínio.

---------------------------------------------------------------------------
The company’s website development team requested the creation of two new numeric columns in the database. On the property site, they would like to present more summarized information regarding values. For this, these columns will be important:


*   valor_por_mes: this column should contain the monthly expenses of each property, including rent and condominium fees;
*   valor_por_ano: this column should contain the annual expenses per property, that is, property tax (IPTU) plus 12 months of rent and condominium fees.





Criando colunas numéricas e importando a base de dados completa:

Creating Numeric Columns and Importing the Complete Database:

```python
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv'
pd.read_csv(url, sep=';')
dados.head()
```

Criando coluna Valor_por_mes:

Creating valor_por_mes Column:

```python
dados['Valor_por_mes'] = dados['Valor'] + dados['Condominio'] #soma dos registros em 'Valor' + 'Condomonio'. Criando a coluna Valor_por_mes e atribuindo o valor da soma de cada registro
dados.head()
```

Criando coluna Valor_por_ano:

Creating valor_por_ano Column:

```python
dados['Valor_por_ano'] = dados['Valor_por_mes'] * 12 + dados['IPTU']
dados.head()
```

O time de desenvolvimento solicitou a criação de mais duas colunas. No entanto, dessa vez elas são categóricas:

*   Descrição: essa coluna deve possuir uma sumarização das principais informações dos imóveis que serão apresentadas no site: tipo de imóvel, bairro, quantidade de quartos e vagas de garagem;
*   Possui_suite: essa deve ser uma coluna que informe apenas se o imóvel possui ou não suítes, sem se importar com a quantidade.

---------------------------------------------------------------------------

The development team requested the creation of two more columns. However, this time they are categorical:

*   Descrição: this column should contain a summary of the main information about the properties that will be presented on the site: property type, neighborhood, number of bedrooms, and parking spaces;
*   Possui_suite: this should be a column that indicates whether the property has suites or not, regardless of the quantity.




Criando coluna Descrição:

Creating Descrição Column:

```python
#astype(str) faz a conversão dos números int para str. Permitindo a concatenação
dados['Descricao'] = dados['Tipo'] + ' em ' + dados['Bairro'] + ' com ' + \
                                        dados['Quartos'].astype(str) + ' quarto(s) ' + \
                                        ' e ' + dados['Vagas'].astype(str) + ' vaga(s) de garagem.'
dados.head()
```

Criando coluna Possui_suite:

Creating Possui_suite Column:

```python
#a função lambda vai fazer a verificação em cada registro de 'Suites' atribuindo "Sim" para as que tiverem Suites > 0 e "Não" para o resto
dados['Possui_suite'] = dados['Suites'].apply(lambda x: "Sim" if x > 0 else "Não")
dados.head()
```

Saving the DataFrame:
